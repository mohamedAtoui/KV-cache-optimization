```python
# === Colab Setup ===
!pip install -q flash-linear-attention transformers datasets peft accelerate matplotlib tqdm

import os
if not os.path.exists('PROJECT'):
    !git clone https://gitlab.cim.rhul.ac.uk/wmis066/PROJECT.git
%cd PROJECT
!pip install -e .

# Mount Google Drive for checkpoints
from google.colab import drive
drive.mount('/content/drive')
CHECKPOINT_DIR = '/content/drive/MyDrive/kv2state_checkpoints'

# Verify kv2state is importable
import kv2state
print(f"kv2state v{kv2state.__version__} loaded successfully")

import sys
import json
import logging
import torch
import numpy as np
import matplotlib.pyplot as plt
from transformers import AutoModelForCausalLM, AutoTokenizer

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger('kv2state')

# Check GPU
print(f"PyTorch: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name()}")
    mem = torch.cuda.get_device_properties(0).total_memory
    print(f"VRAM: {mem / 1e9:.1f} GB")

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {DEVICE}")
```

## Cell 2: Load Llama-3.1-8B-Instruct

```python
MODEL_NAME = "meta-llama/Llama-3.1-8B-Instruct"

print(f"Loading tokenizer: {MODEL_NAME}")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print(f"Loading model: {MODEL_NAME} (bfloat16)")
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.bfloat16,
    device_map='auto',
    attn_implementation='eager',  # Need eager attention to intercept forward pass
)
model.eval()

# Print model config
config = model.config
print(f"\nModel config:")
print(f"  Layers: {config.num_hidden_layers}")
print(f"  Q heads: {config.num_attention_heads}")
print(f"  KV heads: {config.num_key_value_heads}")
print(f"  Head dim: {config.hidden_size // config.num_attention_heads}")
print(f"  Hidden size: {config.hidden_size}")
print(f"  GQA ratio: {config.num_attention_heads // config.num_key_value_heads} Q per KV head")

# Memory usage
param_bytes = sum(p.numel() * p.element_size() for p in model.parameters())
print(f"  Model memory: {param_bytes / 1e9:.2f} GB")
```

## Cell 3: Baseline Perplexity (Full KV Cache)

```python
from kv2state.eval_perplexity import evaluate_perplexity

print("="*60)
print("BASELINE: Full KV cache perplexity on WikiText-2")
print("="*60)

baseline_result = evaluate_perplexity(
    model=model,
    tokenizer=tokenizer,
    dataset_name='wikitext',
    dataset_config='wikitext-2-raw-v1',
    split='test',
    max_length=2048,
    stride=512,
    device=DEVICE,
)

print(f"\n>>> Baseline perplexity: {baseline_result['perplexity']:.2f}")
print(f">>> Average loss: {baseline_result['avg_loss']:.4f}")
print(f">>> Tokens evaluated: {baseline_result['num_tokens']}")
```

## Cell 4: Head Classification

Two options:
- **Option A**: Load DuoAttention pre-computed patterns (preferred, faster)
- **Option B**: Compute attention entropy from scratch (fallback)

```python
from kv2state.head_classifier import load_duo_attention_patterns, compute_attention_entropy, HeadClassification

# --- Option A: DuoAttention patterns ---
DUO_ATTENTION_DIR = None  # Set to path after cloning DuoAttention

# Uncomment to clone DuoAttention and load patterns:
# !git clone https://github.com/mit-han-lab/duo-attention.git duo-attention
# DUO_ATTENTION_DIR = 'duo-attention/attn_patterns/Meta-Llama-3.1-8B-Instruct'

if DUO_ATTENTION_DIR and os.path.exists(DUO_ATTENTION_DIR):
    print("Loading DuoAttention pre-computed head patterns...")
    head_classification = load_duo_attention_patterns(
        DUO_ATTENTION_DIR,
        sparsity=0.5,  # 50% of heads are streaming (convertible)
    )
    classification_method = "DuoAttention"

# --- Option B: Entropy-based classification ---
else:
    print("DuoAttention patterns not found. Using entropy-based classification...")
    print("(This requires a forward pass with output_attentions=True, which is slow)")
    
    # Create a small dataloader for entropy computation
    from datasets import load_dataset
    from torch.utils.data import DataLoader
    
    entropy_dataset = load_dataset('wikitext', 'wikitext-2-raw-v1', split='train')
    # Filter out empty strings
    entropy_dataset = entropy_dataset.filter(lambda x: len(x['text'].strip()) > 50)
    
    def tokenize_fn(examples):
        return tokenizer(
            examples['text'],
            truncation=True,
            max_length=512,  # Short sequences for entropy estimation
            padding='max_length',
            return_tensors='pt',
        )
    
    entropy_dataset = entropy_dataset.map(tokenize_fn, batched=True, remove_columns=['text'])
    entropy_dataset.set_format('torch')
    entropy_loader = DataLoader(entropy_dataset, batch_size=2, shuffle=False)
    
    head_classification = compute_attention_entropy(
        model=model,
        dataloader=entropy_loader,
        max_batches=20,  # Small number for speed
        device=DEVICE,
    )
    classification_method = "Entropy"

# --- Summary ---
print(f"\nClassification method: {classification_method}")
print(f"Mask shape: {head_classification.mask.shape} (layers x KV heads)")
print(f"Retrieval heads: {head_classification.num_retrieval}")
print(f"Streaming heads: {head_classification.num_streaming}")
print(f"Streaming fraction: {head_classification.streaming_fraction:.1%}")

# Show per-layer breakdown
mask = head_classification.mask
print(f"\nPer-layer breakdown (R=retrieval, S=streaming):")
for layer_idx in range(mask.shape[0]):
    heads = ''.join(['R' if mask[layer_idx, h] else 'S' for h in range(mask.shape[1])])
    n_streaming = (~mask[layer_idx]).sum().item()
    print(f"  Layer {layer_idx:2d}: [{heads}] ({n_streaming}/{mask.shape[1]} streaming)")
```

## Cell 5: Apply KV2State Monkey Patch

```python
from kv2state.hybrid_attention import patch_model_for_kv2state, KV2StateConfig

# Configure state conversion
state_config = KV2StateConfig(
    decay_init=0.99,
    learnable_decay=False,  # Fixed decay for zero-shot test
    chunk_size=64,
)

print("Applying KV2State monkey patch...")
print(f"  Decay factor: {state_config.decay_init}")
print(f"  Chunk size: {state_config.chunk_size}")

model, state_cache = patch_model_for_kv2state(
    model=model,
    head_classification=head_classification,
    config=state_config,
)

print(f"\nPatch applied successfully!")
print(f"State cache initialized. Memory per batch: {state_cache.memory_bytes / 1024:.1f} KB")
```

## Cell 6: Zero-Shot Perplexity (State-Converted Streaming Heads)

```python
print("="*60)
print("KV2STATE: Zero-shot perplexity with state-converted streaming heads")
print("="*60)

# Reset state cache before evaluation
state_cache.reset()

kv2state_result = evaluate_perplexity(
    model=model,
    tokenizer=tokenizer,
    dataset_name='wikitext',
    dataset_config='wikitext-2-raw-v1',
    split='test',
    max_length=2048,
    stride=512,
    device=DEVICE,
)

ratio = kv2state_result['perplexity'] / baseline_result['perplexity']

print(f"\n{'='*60}")
print(f"RESULTS COMPARISON")
print(f"{'='*60}")
print(f"Baseline perplexity:  {baseline_result['perplexity']:.2f}")
print(f"KV2State perplexity:  {kv2state_result['perplexity']:.2f}")
print(f"Degradation ratio:    {ratio:.2f}x")
print(f"")

# GO/NO-GO decision
if ratio < 3.0:
    print(">>> GO: Perplexity degradation < 3x. Proceed to calibration (Sprint 2).")
    print(">>> The idea works! State conversion captures enough information.")
elif ratio < 10.0:
    print(">>> MARGINAL: Perplexity degradation between 3-10x.")
    print(">>> Consider adjusting decay factor, trying GLA/DeltaNet, or reducing conversion fraction.")
else:
    print(">>> STOP: Perplexity degradation > 10x. Fundamental issue with approach.")
    print(">>> Need to re-evaluate: check feature maps, state update rule, or head classification.")
```

## Cell 7: Decay Factor Sweep

```python
import importlib
import kv2state.hybrid_attention as ha_module

decay_values = [0.9, 0.95, 0.99, 0.999]
decay_results = {}

print("Sweeping decay factor values...")
print(f"Baseline perplexity: {baseline_result['perplexity']:.2f}")
print()

for decay in decay_values:
    print(f"--- Testing decay = {decay} ---")
    
    # Reload model fresh for each decay value
    # (In practice, we'd just update the decay parameter, but for zero-shot
    #  the cleanest approach is to re-patch)
    
    # Update decay in existing state modules
    import math
    log_decay = math.log(decay / (1.0 - decay))
    for name, mod in model._kv2state_modules.items():
        mod.log_decay.data.fill_(log_decay)
    
    state_cache.reset()
    
    result = evaluate_perplexity(
        model=model,
        tokenizer=tokenizer,
        dataset_name='wikitext',
        dataset_config='wikitext-2-raw-v1',
        split='test',
        max_length=2048,
        stride=512,
        device=DEVICE,
    )
    
    ratio = result['perplexity'] / baseline_result['perplexity']
    decay_results[decay] = result
    
    print(f"  Perplexity: {result['perplexity']:.2f} ({ratio:.2f}x baseline)")
    print()

# Summary table
print("\n" + "="*60)
print(f"{'Decay':<10} {'Perplexity':<15} {'Ratio':<10} {'Status':<15}")
print("="*60)
for decay, result in decay_results.items():
    ratio = result['perplexity'] / baseline_result['perplexity']
    status = 'GO' if ratio < 3.0 else ('MARGINAL' if ratio < 10.0 else 'STOP')
    print(f"{decay:<10.3f} {result['perplexity']:<15.2f} {ratio:<10.2f} {status:<15}")
print(f"{'Baseline':<10} {baseline_result['perplexity']:<15.2f} {'1.00':<10} {'Reference':<15}")
```

## Cell 8: Visualize Head Classification & Entropy

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# --- Plot 1: Head classification heatmap ---
ax1 = axes[0]
mask_np = head_classification.mask.numpy().astype(float)
im = ax1.imshow(mask_np, aspect='auto', cmap='RdYlGn', interpolation='nearest')
ax1.set_xlabel('KV Head Index')
ax1.set_ylabel('Layer Index')
ax1.set_title('Head Classification\n(Green=Retrieval, Red=Streaming)')
ax1.set_xticks(range(mask_np.shape[1]))
plt.colorbar(im, ax=ax1, label='1=Retrieval, 0=Streaming')

# --- Plot 2: Raw scores heatmap ---
ax2 = axes[1]
if head_classification.raw_scores is not None:
    scores_np = head_classification.raw_scores.numpy()
    im2 = ax2.imshow(scores_np, aspect='auto', cmap='viridis', interpolation='nearest')
    ax2.set_xlabel('KV Head Index')
    ax2.set_ylabel('Layer Index')
    ax2.set_title(f'Raw Scores ({classification_method})')
    ax2.set_xticks(range(scores_np.shape[1]))
    plt.colorbar(im2, ax=ax2, label='Score')
else:
    ax2.text(0.5, 0.5, 'No raw scores available', ha='center', va='center', transform=ax2.transAxes)

# --- Plot 3: Score distribution (bimodality) ---
ax3 = axes[2]
if head_classification.raw_scores is not None:
    flat_scores = head_classification.raw_scores.numpy().flatten()
    ax3.hist(flat_scores, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
    ax3.axvline(x=0.5, color='red', linestyle='--', label='Default threshold')
    ax3.set_xlabel('Score')
    ax3.set_ylabel('Count')
    ax3.set_title('Score Distribution\n(Bimodal = good separation)')
    ax3.legend()
else:
    ax3.text(0.5, 0.5, 'No raw scores available', ha='center', va='center', transform=ax3.transAxes)

plt.tight_layout()
plt.savefig('head_classification.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: head_classification.png")

# --- Plot 4: Decay sweep results ---
if decay_results:
    fig2, (ax4, ax5) = plt.subplots(1, 2, figsize=(12, 5))
    
    decays = list(decay_results.keys())
    ppls = [r['perplexity'] for r in decay_results.values()]
    ratios = [r['perplexity'] / baseline_result['perplexity'] for r in decay_results.values()]
    
    ax4.plot(decays, ppls, 'bo-', markersize=8, linewidth=2)
    ax4.axhline(y=baseline_result['perplexity'], color='green', linestyle='--', label='Baseline')
    ax4.axhline(y=3*baseline_result['perplexity'], color='orange', linestyle='--', label='3x (GO threshold)')
    ax4.set_xlabel('Decay Factor (lambda)')
    ax4.set_ylabel('Perplexity')
    ax4.set_title('Perplexity vs Decay Factor')
    ax4.legend()
    ax4.set_xscale('linear')
    
    ax5.bar(range(len(decays)), ratios, color=['green' if r < 3 else 'orange' if r < 10 else 'red' for r in ratios])
    ax5.set_xticks(range(len(decays)))
    ax5.set_xticklabels([str(d) for d in decays])
    ax5.axhline(y=3.0, color='orange', linestyle='--', label='GO threshold (3x)')
    ax5.axhline(y=10.0, color='red', linestyle='--', label='STOP threshold (10x)')
    ax5.set_xlabel('Decay Factor')
    ax5.set_ylabel('Perplexity Ratio (vs baseline)')
    ax5.set_title('Degradation Ratio by Decay Factor')
    ax5.legend()
    
    plt.tight_layout()
    plt.savefig('decay_sweep.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("Saved: decay_sweep.png")
```
